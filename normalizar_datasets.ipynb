```python
import pandas as pd
import numpy as np
```

## Scopus

```python
df_scopus_cf = pd.read_csv('archivos/scopus_estimation_carbon_footprint.csv')

df_scopus_cf
```

```python
df_scopus_cf.dtypes
```

```python
df_scopus_cf = df_scopus_cf[['Title', 'Document Type', 'Authors', 'Source title', 'Volume', 'Year', 'Page start', 'Page end', 
                             'DOI', 'Link', 'ISSN', 'Source', 'Author Keywords', 'Abstract']]

df_scopus_cf.rename(columns={'Author Keywords' : 'Keywords'}, inplace=True)

df_scopus_cf['Volume'] = df_scopus_cf['Volume'].str.replace(r'\D', '', regex=True)
df_scopus_cf['Volume'].replace('', np.nan, inplace=True)
df_scopus_cf['Volume'] = df_scopus_cf['Volume'].astype('float64')

df_scopus_cf
```

```python
df_scopus_cf.dtypes
```

## Sciencedirect

```python
df_sciencedirect_cf = pd.read_csv('archivos/sciencedirect_carbon_footprint.csv')

df_sciencedirect_cf
```

```python
df_sciencedirect_cf.dtypes
```

```python
def normalizar_listas(valor):
    return str(valor).replace('[', '').replace(']', '').replace("'", '').replace(',', ';')

df_sciencedirect_cf.rename(columns={'authors' : 'Authors', 'primary_title' : 'Title', 'doi' : 'DOI', 'issn' : 'ISSN', 
                                       'year' : 'Year', 'journal_name' : 'Source title', 'volume' : 'Volume',
                                       'start_page' : 'Page start', 'end_page' : 'Page end', 'keywords' : 'Keywords',
                                       'urls' : 'Link', 'abstract' : 'Abstract', 'type_of_reference' : 'Document Type'}, inplace=True)

df_sciencedirect_cf['Authors'] = df_sciencedirect_cf['Authors'].apply(normalizar_listas)
df_sciencedirect_cf['Keywords'] = df_sciencedirect_cf['Keywords'].apply(normalizar_listas)
df_sciencedirect_cf['Source'] = df_sciencedirect_cf['Authors'].apply(lambda x: 'Sciencedirect')
df_sciencedirect_cf['Link'] = df_sciencedirect_cf['Link'].apply(normalizar_listas)
df_sciencedirect_cf['Document Type'] = df_sciencedirect_cf['Document Type'].replace({'JOUR' : 'Journal'})

df_sciencedirect_cf.drop(columns=['date', 'secondary_title', 'number'], inplace=True)

df_sciencedirect_cf
```

```python
df_sciencedirect_cf.dtypes
```

```python
df_unido = pd.concat([df_scopus_cf, df_sciencedirect_cf])

df_unido.to_csv('archivos/dataset_unificado.csv', sep=',')

df_unido
```
