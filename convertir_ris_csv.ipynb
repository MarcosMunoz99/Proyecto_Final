```python
import rispy
import pandas as pd 
```

```python
lista_archivos = ['sciencedirect_carbon_footprint_01', 'sciencedirect_carbon_footprint_02', 'sciencedirect_carbon_footprint_03']
lista_dfs = []

for i in range(len(lista_archivos)):
    archivo = lista_archivos[i]

    # Carga el archivo RIS
    with open(f'archivos/{archivo}.ris', 'r') as file:
        entries = rispy.load(file)

    # Convierte a DataFrame de pandas
    df = pd.DataFrame(entries)

    df.to_csv(f'archivos/{archivo}.csv', sep=',', index=False)

    lista_dfs.append(df)
```

```python
df_unido = pd.concat(lista_dfs, ignore_index=True)

df_unido.to_csv('archivos/sciencedirect_carbon_footprint.csv', sep=',', index=False)

df_unido
```
